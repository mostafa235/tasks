```python
import pandas as pd
```

```python
data=pd.read_csv('dastasets/visitor-interests.csv')
```

```python
data=data[:51]
```

```python
import user_agents 
def extract(ua):
    ua=user_agents.parse(ua)
    return pd.Series((ua.os.family,ua.browser.family,ua.device.family))
data[['os','browser','device']]=data['UserAgent'].apply(extract)
data.head()
```

```python
data=data[:51]
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
def Ip_extract(ip):
    try:   
        ip=ip2geo.get(ip,api_key='free')
        return pd.Series((ip.latitude,ip.longitude))

    except:
        return 'missing'

data[['latitude','longitude']]=data['IP'].apply(Ip_extract)
data.head()
```
