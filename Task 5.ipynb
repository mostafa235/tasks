```python
import pandas as np 
```

```python
data=np.read_csv('dastasets/visitor-interests.csv')
```

```python
data.head()
```

```python
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
data=data[:10]
```

```python
def extract_country_from_ip(ip):
    try:
        response = ip2geo.get(ip, api_key='free')
        return response.country
    except:
        return 'Missing'

data['latitude']=data['IP'].apply(extract_country_from_ip)
data.head()
```
