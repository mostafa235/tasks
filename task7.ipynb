```python
data.dropna(axis=0)
data.dropna(axis=1)
```

```python
numerical_colms = ['Id',  'TotalPay', 'TotalPayBenefits', 'Year']

for col in numerical_colms:
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
data.drop(['Status'],axis=1,inplace=True)
```

```python
data.drop(['Notes'],axis=1,inplace=True)
```

```python
data['BasePay'] = data['BasePay'].apply(lambda x: abs(x))
data['OvertimePay'] = data['OvertimePay'].apply(lambda x: abs(x))
data['OtherPay'] = data['OtherPay'].apply(lambda x: abs(x))
data['Benefits'] = data['Benefits'].apply(lambda x: abs(x))
data['TotalPay'] = data['TotalPay'].apply(lambda x: abs(x))
data['TotalPayBenefits'] = data['TotalPayBenefits'].apply(lambda x: abs(x))
```
