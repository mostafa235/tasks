```python

data['not.fully.paid'].isnull().mean()
data.groupby('not.fully.paid')['not.fully.paid'].count()/len(data)
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.countplot(x='not.fully.paid', data=data)
```

```python
count_class_0, count_class_1 = data['not.fully.paid'].value_counts()
data_0 = data[data['not.fully.paid'] == 0]
data_1 = data[data['not.fully.paid'] == 1]
data_1_over = data_1.sample(count_class_0, replace=True)
data_test_over = pd.concat([data_0, data_1_over], axis=0)
print('Random over-sampling:')
print(data_test_over['not.fully.paid'].value_counts())

sns.set_style('darkgrid')
sns.countplot(x='not.fully.paid', data=data_test_over)
```
